```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle

from collections import deque
```

## Part1: Deep Q-Network

### **Question 1: basic Q-learning performance**

```python
with open('./logs/dqn.pkl', 'rb') as f:
    dqn = pickle.load(f)
```

```python
dqn_best_mean_reward = []
dqn_mean_reward_100  = []
dqn_timestep = []

for log in dqn:
    dqn_best_mean_reward.append(log['best mean reward'])
    dqn_mean_reward_100.append(log['mean reward (100 episodes)'])
    dqn_timestep.append(log['Timestep'] / 10000)
```

```python
plt.title('Pong with DQN-2015')

plt.plot(np.arange(len(dqn)), dqn_mean_reward_100, label = 'DQN')
#plt.plot(np.arange(len(logs)), best_mean_reward, label = 'max rewards')

plt.legend()
plt.xticks(np.arange(0,dqn_timestep[-1],200))
plt.xlabel('TimeSteps(10k)')
plt.ylabel('average rewards')
plt.show()
```

### **Question 2: DQN vs Double DQN**

```python
with open('./logs/ad5244aa-8a4b-453a-9a5e-25f65c295e2e.pkl', 'rb') as f:
    ddqn = pickle.load(f)
```

```python
ddqn_best_mean_reward = []
ddqn_mean_reward_100  = []
ddqn_timestep = []

for log in ddqn:
    ddqn_best_mean_reward.append(log['best mean reward'])
    ddqn_mean_reward_100.append(log['mean reward (100 episodes)'])
    ddqn_timestep.append(log['Timestep'] / 10000)
```

```python
plt.title('DQN vs DDQN')

plt.plot(np.arange(len(ddqn)), ddqn_mean_reward_100, label = 'DDQN')
plt.plot(np.arange(len(ddqn)), dqn_mean_reward_100[:len(ddqn)], label = 'DQN')

plt.legend()
plt.xticks(np.arange(0,ddqn_timestep[-1],200))
plt.xlabel('TimeSteps(10k)')
plt.ylabel('average rewards')
plt.show()
```

***Brief Analysis***

*Double Deep Q-network has shown slightly better performance than Deep Q-network in the earlier stage of the training. However, both of the networks eventually converged to the score between 15~20.*
