***Brief Analysis***

*1. Reward to Go allows agent to hit maximum rewards. Without reward-to-go, agent seems to concentrate on earlier behvaiors, rather decreases the continuity.*<br>
*2. Normalizing Advantageous Function stabilizes the training with reducing the high variance in Advantageous Function. Even though the agent reaches the maximum score, raw advantageous function disturbs agent from the convergence*<br>
*3. Since large batch size reduces the high variance, it helps agent's policy to stabilize. However, it doesn't seem to aid the speed of the convergence in this test case*
